```python
import pandas as pd
```

```python
df = pd.read_csv('Telco Customer Churn.csv')
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.tail()
```

```python
df.nunique()
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
df[cat_cols].nunique()
```

```python
df = df.drop(['customerID'], axis=1)
cat_cols.remove('customerID')
```

```python
df.head()
```

```python
df[cat_cols].nunique()
```

```python
for col in cat_cols:
  print(f'{col}: {df[col].unique()}')
```

```python
df['gender'] = df['gender'].replace({'Male': 0, 'Female': 1})
df['Partner'] = df['Partner'].replace({'Yes': 1, 'No': 0})
df['Dependents'] = df['Dependents'].replace({'Yes': 1, 'No': 0})
df['PhoneService'] = df['PhoneService'].replace({'Yes': 1, 'No': 0})
df['PaperlessBilling'] = df['PaperlessBilling'].replace({'Yes': 1, 'No': 0})
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 0, 'No': 1, 'Yes': 2})
df['InternetService'] = df['InternetService'].replace({'DSL': 1, 'Fiber optic': 2, 'No': 0})
df['OnlineSecurity'] = df['OnlineSecurity'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['OnlineBackup'] = df['OnlineBackup'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['DeviceProtection'] = df['DeviceProtection'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['TechSupport'] = df['TechSupport'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['StreamingTV'] = df['StreamingTV'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['StreamingMovies'] = df['StreamingMovies'].replace({'No': 1, 'Yes': 2, 'No internet service': 0})
df['Contract'] = df['Contract'].replace({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
df['PaymentMethod'] = df['PaymentMethod'].replace({'Electronic check': 0, 'Mailed check': 1, 'Bank transfer (automatic)': 2, 'Credit card (automatic)': 3})
```

```python
df.head()
```

```python
df.dtypes
```

```python
X = df.drop(columns='Churn')
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=786)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
dtc_prediction = dtc.predict(X_test)
```

```python
dtc_prediction
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
dtc_accuracy = accuracy_score(y_test, dtc_prediction)
dtc_accuracy
```

```python
confusion_matrix(y_test, dtc_prediction)
```

```python
print(classification_report(y_test, dtc_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=10)
rfc.fit(X_train, y_train)
```

```python
rfc_prediction = rfc.predict(X_test)
```

```python
rfc_accuracy = accuracy_score(y_test, rfc_prediction)
rfc_accuracy
```

```python
confusion_matrix(y_test, rfc_prediction)
```

```python
print(classification_report(y_test, rfc_prediction))
```
